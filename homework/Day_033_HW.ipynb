### 1. 模型的泛化能力 (generalization) 是指什麼？

    當該模型可以對於某特定sampling data有非常好的表現，但在其他的sampling data中的預測表現很差，該模型則是過度適應(over-fitting)，
    泛化能力較差。
    如模型在許多不同的sampling data中的能有不錯的表現，不會因為新資料的加入而導致預測表現變差，則稱此模型的泛化能力較好。
    泛化能力用於描述機器學習出來的模型在預測問題時，是否能彈性地推廣至不同情境的能力。

### 2. 分類問題與回歸問題分別可用的目標函數有哪些？

    分類問題使用交叉熵(cross entropy)，
    回歸問題使用均方誤差(Mean square error，MSE)和平均絕對值誤差(Mean absolute error，MAE)。
