# 作業
* 在精簡深度學習的方式上 : 卷積類神經 (CNN) 採用像素遠近，而遞歸類神經 (RNN) 採用著則是時間遠近
* 那麼，既然有著類似的設計精神，兩者是否有可能互換應用呢?

CNN和RNN可以互換應用，這取決於以甚麼樣的角度去思考問題。
例如電磁訊號的分析，因為電磁訊號的輸入是有時間序列性的，使用RNN是非常直覺的。
但是如果在意的只是特定的訊號波型是否出現，透過專門處理影像像素的CNN來處理問題，可能會更恰當。

另外通常CNN會用來處理影片畫面的分析中，而若問題是更重視時間序列的特性，如因果關係，搭配使用RNN會更能貼近問題的核心。
CNN和RNN也可以搭配交互使用，端視需要處理的問題是甚麼。
